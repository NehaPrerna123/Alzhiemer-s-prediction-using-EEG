```python
import os
import pandas as pd

# Path to the folder containing CSV files
folder_path = r'D:\RESEARCH 2024\ALZEIMER II\HC'  # Update if needed

# List to hold processed DataFrames
dataframes = []

# Number of rows to pick from each CSV
num_rows = 1000

for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)

        # Drop 'Time' column if it exists
        if 'Time' in df.columns:
            df = df.drop(columns=['Time'])

        # Calculate start and end indices to select 1000 samples from the middle
        total_rows = len(df)
        if total_rows >= num_rows:
            start = (total_rows - num_rows) // 2
            df = df.iloc[start:start + num_rows]
        else:
            print(f"Warning: {filename} has less than {num_rows} rows ({total_rows} rows). Skipping.")
            continue

        # Add 'target' column
        df['target'] = 'HC'

        dataframes.append(df)

# Concatenate and save
if dataframes:
    merged_df = pd.concat(dataframes, ignore_index=True)
    output_file = os.path.join(folder_path, 'merged_HC.csv')
    merged_df.to_csv(output_file, index=False)
    print(f"Merged dataset saved at:\n{output_file}")
else:
    print("No valid files found or all files have fewer than 1000 rows.")
```
