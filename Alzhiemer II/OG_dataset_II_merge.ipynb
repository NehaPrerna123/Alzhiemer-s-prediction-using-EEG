```python
import pandas as pd

# Load CSV files
ai_file = 'D:\\RESEARCH 2024\\ALZEIMER II\\Merged data\\AL_merged.csv'
mci_file = 'D:\\RESEARCH 2024\\ALZEIMER II\\Merged data\\MCI_merged.csv'
hc_file = 'D:\\RESEARCH 2024\\ALZEIMER II\\Merged data\\HC_merged.csv'

# Read each dataset (assuming they already include a 'label' column)
df_ai = pd.read_csv(ai_file)
df_mci = pd.read_csv(mci_file)
df_hc = pd.read_csv(hc_file)

# Check if label column exists in all
assert 'target' in df_ai.columns, "Label column missing in AL_merged.csv"
assert 'target' in df_mci.columns, "Label column missing in MCI_merged.csv"
assert 'target' in df_hc.columns, "Label column missing in HC_merged.csv"

# Merge the datasets
df_merged = pd.concat([df_ai, df_mci, df_hc], axis=0, ignore_index=True)

# Optional: Shuffle the combined dataset
df_merged = df_merged.sample(frac=1).reset_index(drop=True)

# Save merged dataset
output_path = 'D:\\RESEARCH 2024\\ALZEIMER II\\Merged data\\OG_AL_II.csv'
df_merged.to_csv(output_path, index=False)

print(f"Merged dataset saved at:\n{output_path}")

```
